```python
import numpy
import matplotlib.pyplot as plt
import pandas

# CF Read in my csv file from SDSS Query
df = pandas.read_csv('sdss_search.csv', skiprows = 1)
print(df)

# CF Get RA and DEC from that query

ra = df.iloc[:, 0]
dec = df.iloc[:, 1]

# CF Plot RA vs. Dec

plt.figure(figsize = (8,8))
plt.scatter(ra, dec)
plt.xlabel("Right Ascension")
plt.ylabel("Declination")
plt.show()

# CF Make it so larger g magnitudes will be plotted with smaller circles 

g_area = (((1 / df.iloc[:, 2]) * 225) ** 2)

# CF Make the plot of RA and DEC with larger g magnitudes being plotted with smaller circles
plt.figure(figsize = (8,8))
plt.scatter(df.iloc[:, 0],df.iloc[:, 1], s = g_area)
plt.xlabel("Right Ascension")
plt.ylabel("Declination")
plt.show()
```
